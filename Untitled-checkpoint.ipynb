```python
feed_back=data = pd.read_csv(r"C:\Users\Virein Harjani\OneDrive\Documents\csv\Survey data _Customer comments.csv")
```

```python
feed_back.head(20)
```

```python
comments = feed_back['verbatim_text']
```

```python
def preprocess_text(text):
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.isalpha()]
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    return tokens
preprocessed_comments = comments.apply(preprocess_text)
```

```python
print(preprocessed_comments)
```

```python
all_tokens = [token for tokens in preprocessed_comments for token in tokens]
```

```python
freq_dist = FreqDist(all_tokens)

# Get most common words/themes
common_words = freq_dist.most_common(100)
print("Most common words:", common_words)
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_dist)

# Display the WordCloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
sia = SentimentIntensityAnalyzer()
sentiments = comments.apply(lambda x: sia.polarity_scores(x))

# Extract themes using predefined keywords
themes = {
    'quality': ['quality', 'taste', 'flavor'],
    'service': ['service', 'wait', 'staff''fast','slow'],
    'variety': ['variety', 'options', 'choices','more','less'],
    'food':['cream','salad','snack','coffee','meals','wine','fruit','chicken','water','cookies','dinner'],
    # Add more themes and keywords as needed
}

# Function to categorize comments into themes
def categorize_comment(comment):
    for theme, keywords in themes.items():
        for keyword in keywords:
            if keyword in comment:
                return theme
    return 'other'

# Categorize comments into themes
data['theme'] = comments.apply(categorize_comment)

# Plot bar chart of theme frequencies
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='theme', order=data['theme'].value_counts().index)
plt.title('Frequent Themes in Survey Comments')
plt.xlabel('Themes')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
